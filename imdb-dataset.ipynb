```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Creates the IMDB data frame from the given 'movie_metadata.csv' 
imdb_df = pd.read_csv('movie_metadata.csv')


# Data cleaning
imdb_df = imdb_df.dropna()
imdb_df = imdb_df.drop_duplicates()

# ------------------------------------------------------------------------------------
# Q1 - Successful actors and directors
# ------------------------------------------------------------------------------------

# Find successful actors
def successfulActors():
    try:
        number_of_actors = int(input('\nEnter the number of actors you want to show: '))
        # Data cleaning
    
        # 1. Create three actors dataframe and rename the column name to be same    
        actor1_df = imdb_df[['actor_1_name', 'gross']]
        actor1_df = actor1_df.rename(columns = {"actor_1_name": "actor_name", "gross": "gross"})

        actor2_df = imdb_df[['actor_2_name', 'gross']]
        actor2_df = actor2_df.rename(columns = {"actor_2_name": "actor_name", "gross": "gross"})

        actor3_df = imdb_df[['actor_3_name', 'gross']]
        actor3_df = actor3_df.rename(columns = {"actor_3_name": "actor_name", "gross": "gross"})

        # 2. Merge three separate actors column into one    
        total_actors_df = pd.concat([actor1_df, actor2_df, actor3_df]).groupby('actor_name', as_index = False).sum()

        # Check if the value is less than or equals to the total number of actors dataset
        if number_of_actors > len(total_actors_df):
            print(f'\nPlease try again. Total actors are only {len(total_actors_df)}')
            successfulActors()
            return

        # 3. Sort the data frame by gross column in descending values    
        sorted_gross_df = total_actors_df.sort_values(by = ['gross'], ascending = False)

        # 4. Read the actor column from the sorted data frame    
        actors_df = sorted_gross_df[['actor_name', 'gross']]

        # 5. Print the successful directors
        print(actors_df.head(number_of_actors))

        # 6. Show the successful actors graph 
        sns.barplot(data = sorted_gross_df.head(number_of_actors), x = 'gross', y = 'actor_name')
        plt.ylabel('Actors')
        plt.xlabel('Gross Income')
        plt.title(f'Top {number_of_actors} successful actors')
        plt.show()
        
    except:
        print('Invalid input. Please try again.')
        successfulActors()
        

# Find successful directors
def successfulDirectors():
    try:
        number_of_directors = int(input('\nEnter the number of directors you want to show: '))
        # Data Cleaning
    
        # 1. Find the unique directors detail 
        total_directors_df = imdb_df.groupby(['director_name'], as_index = False).sum()

        # 2. Check if the value is less than or equals to the total number of directors dataset
        if number_of_directors > len(total_directors_df):
            print(f'\nPlease try again. Total directors are only {len(total_directors_df)}')
            successfulActors()
            return

        # 3. Sort the data frame by gross column in descending values    
        sorted_df = total_directors_df.sort_values(by = ['gross'], ascending = False)

        # 4. Read the director column from the sorted data frame
        direct_df = sorted_df[['director_name', 'gross']]

        # 5. Print the successful directors
        print(direct_df.head(number_of_directors))

        # 6. Show the successful directors graph
        sns.barplot(data = sorted_df.head(number_of_directors), y = 'director_name', x = 'gross')   
        plt.ylabel('Directors')
        plt.xlabel('Gross Income')
        plt.title(f'Top {number_of_directors} successful directors')
        plt.xticks(rotation=90)
        plt.tight_layout()

        plt.show()
        
    except:
        print('Invalid input. Please try again.')
        successfulDirectors()        

# Find most successful directors or actors
def findSuccessfulActorsAndDirectors():
    
    print(' i. Top Directors')
    print('ii. Top Actors')
    
    # 1. Take the user input
    user_option = input('\nPlease choose an option: ')

    if user_option == 'i':
        successfulDirectors()
    elif user_option == 'ii':
        successfulActors()
    else:
        print('\nInvalid Option! Please try again.')
        findSuccessfulActorsAndDirectors()            
# ------------------------------------------------------------------------------------



# ------------------------------------------------------------------------------------
# Q2 Film comparison
# ------------------------------------------------------------------------------------

# Strip the space from front and end of the movie name if any.
all_films = imdb_df['movie_title'].str.lstrip().str.rstrip().str.lower()

def filmComparison():
    
    # 1. Get the first film from the user
    first_film = getFirstFilm()
    
    # 2. Get the second film from the user
    second_film = getSecondFilm()
        
    # User now have correct film names entered by the user
    # Ask for more info
    compareFilmDetails(first_film, second_film)

        
def compareFilmDetails(film1, film2):
    print()
    print('  i. IMDB Scores')
    print(' ii. Gross Earning')
    print('iii. Movie Facebook Likes')
    
    user_option = input('\nPlease choose an option: ')
    
    # 1. Bool data frame for first film
    first_film_df = all_films == film1.lower()
    
    # 2. Bool data frame for second film
    second_film_df = all_films == film2.lower()
    
    # 3. Get the combined dataframe for first film and second film 
    film_compare_df = imdb_df[first_film_df | second_film_df]
    comparison_df = film_compare_df[['gross', 'imdb_score', 'movie_facebook_likes']]
    # print(f'--------------\n {comparison_df}')
    
    if user_option == 'i':
        # Compare IMDB Scores
        film_compare_df.plot.bar(x = 'movie_title', y = 'imdb_score')
        plt.xlabel('Movie Title')
        plt.ylabel('IMDB Scores')
        plt.legend(['IMDB Scores'])
        plt.xticks(rotation=30)
        plt.show()
                
    elif user_option == 'ii':
        # Compare gross earnings
        film_compare_df.plot.bar(x = 'movie_title', y = 'gross')
        plt.xlabel('Movie Title')
        plt.ylabel('Gross Earning')
        plt.legend(['Gross Earning'])
        plt.xticks(rotation=30)
        plt.show()
        
    elif user_option == 'iii':
        # Compare movie facebook likes
        film_compare_df.plot.bar(x = 'movie_title', y = 'movie_facebook_likes')
        plt.xlabel('Movie Title')
        plt.ylabel('Movie Facebook Likes')
        plt.legend(['Movie Facebook Likes'])
        plt.xticks(rotation=30)
        plt.show()
        
    else:
        print('\nInvalid Option! Please try again.')
        compareFilmDetails(film1, film2)
      
        
def getFirstFilm():
    user_continue = True
    
    while user_continue:        
        # Ask user to enter the first film
        first_film = input('Enter 1st film: ')    

        # Check if film is available in the dataset 
        # Otherwise ask user to enter the film name again
        first_film_found = all_films[all_films == first_film.lower()]

        if len(first_film_found) > 0:
            user_continue = False
            return first_film
        else:
            print('\nInvalid film name. Try again.')
        
def getSecondFilm():
    user_continue = True
    
    while user_continue:
        # Ask user to enter the second film
        second_film = input('Enter 2nd film: ')    

        # Check if film is available in the dataset 
        # Otherwise ask user to enter the film name again
        second_film_found = all_films[all_films == second_film.lower()]
                        
        if len(second_film_found) > 0:
            user_continue = False
            return second_film
        else:
            print('\nInvalid film name. Try again.')

            
# ------------------------------------------------------------------------------------
# Q3 Analyse the distribution of gross earnings
# ------------------------------------------------------------------------------------

def analyseGrossEarnings():
    
    # 1. Get the title_year data frame
    year_df = imdb_df['title_year']
    
    max_year = year_df.max()
    min_year = year_df.min()
            
    user_continue = True    
    while user_continue:        
        try:            
            # 2. Get the user input of start year
            start_year = int(input('\nEnter start year: '))

            # 3. Check if the user has entered the start year in the correct range                     
            if start_year < min_year or start_year > max_year:
                print(f'Invalid year. Available range is between {min_year} - {max_year}')
            else:
                user_continue = False
        except:
            print('Invalid year. Year must be a digit.')
            
    
    # Reset the user_continue value
    user_continue = True
    
    while user_continue:
        try:
            # 4. Get the user input of end year
            end_year = int(input('\nEnter end year: '))

            # 5. Check if the user has entered the end year in the correct range
            if end_year < start_year or end_year > max_year:
                print(f'Invalid year. Available range is between {start_year} - {max_year}')
            else:                
                user_continue = False
        except:
            print('Invalid year. Year must be a digit.')
   
    # 6. Filter movies data frame between start and end year
    filtered_movies = imdb_df[(year_df >= start_year) & (year_df <= end_year)]
    
    # 7. Find min, average and max gross of the films in each year between the start and end year
    mean_list = []
    min_list = []
    max_list = []
    year_list = []
    
    for i in range(start_year, end_year + 1):

        # 8. Filter movies for a single year
        film_df = filtered_movies[filtered_movies['title_year'] == i]['gross']        
        
        # 9. Find mean, minimum and maximum gross for the films in that year
        gross_mean = film_df.mean()
        gross_min = film_df.min()
        gross_max = film_df.max()
        
        mean_list.append(gross_mean)
        min_list.append(gross_min)
        max_list.append(gross_max)
        year_list.append(i)
        
        print(f'Movies grossing in year: {i} has max {gross_max}, minimum {gross_min} and average {gross_mean}')
    
    # 10. Create the final dictionary for the data frame
    final_data = {'Minimum': min_list, 'Maximum': max_list, 'Mean': mean_list}
    
    # 11. Create the data frame using the final dictionary with years as the index
    final_df = pd.DataFrame(final_data, index = year_list)
    
    
    # 12. Plot the dataframe
    final_df.plot()
    
    plt.show()

# ------------------------------------------------------------------------------------
# Q4 Genre Analysis
# ------------------------------------------------------------------------------------

def generalAnalysis():
    
    # 1. All genres data frame and create list of all genres 
    all_genres_df = imdb_df['genres'].str.split('|')
    all_genres = []
    for genre in all_genres_df:
        all_genres += genre
        
    # 2. Find all unique genres
    unique_genres = set(all_genres)
    all_genres_string = ' | '.join(unique_genres)

    # 3. Show all the unique genres to the user
    print(f"List of all genres: \n\n{all_genres_string}")

    # 4. Continue asking the user to enter the genre until its correct
    is_correct_genre = False  
    while not is_correct_genre:

        user_genre = input(f'Please enter an genre from the list provided: ').title()

        # 5. Check if user entered the correct genre if not then ask the user to re enter.
        if user_genre.title() in unique_genres:
            
            is_correct_genre = True
            
            # 6. Convert all genres name to lowercase for the comparison
            imdb_df['genres'].str.lower()
            
            # 7. Find the data frame for the user provided genre
            films_by_genre = imdb_df[imdb_df['genres'].str.contains(user_genre)]    
            
            # 8. Get the mean of imdb score for the particular genre
            genre_mean = films_by_genre['imdb_score'].mean()
            
            # print(films_by_genre['genres'])
            print(f'\nMean IMDB score of {user_genre} is {genre_mean}')
        
        else:
            print('Invalid genre. Please enter correct genre.')


# ------------------------------------------------------------------------------------
# Q5 Earnings and IMDB scores
# ------------------------------------------------------------------------------------

def plotEarningsAndIMDBScores():    
    
    imdb_column = ['imdb_score']

    first_row = ['num_critic_for_reviews', 'duration', 'director_facebook_likes', 'num_voted_users']        
    second_row = ['actor_1_facebook_likes', 'actor_2_facebook_likes', 'actor_3_facebook_likes', 'cast_total_facebook_likes']    
    third_row = ['facenumber_in_poster', 'num_user_for_reviews', 'title_year', 'budget']
    fourth_row = ['aspect_ratio', 'movie_facebook_likes', 'gross']

    all_rows = [first_row, second_row, third_row, fourth_row]
    
    for row in all_rows:
        sns.pairplot(imdb_df, x_vars = row, y_vars = imdb_column, height = 5, aspect =.8, kind="reg")
        
    plt.show()
    # NOTE : Uncomment this line to see correlation values
    #print(imdb_df.corr())
    print()
    
# ------------------------------------------------------------------------------------
# Q6 Menu and Exit
# ------------------------------------------------------------------------------------

# Show starting Menu    
def showMenu():
    print('\nPlease select one of the following options:\n')
    print('  1. Most successful directors or actors')
    print('  2. Film comparison')
    print('  3. Analyse the distribution of gross earnings')
    print('  4. Genre Analysis')
    print('  5. Earnings and IMDB scores')
    print('  6. Exit')

    exit = 0
    try:
        exit = int(input("\nPlease choose an option from 1 to 6: "))
    except:
        print('\nInvalid Option! Please try again.\n')
        showMenu()
    
    if exit == 1:
        findSuccessfulActorsAndDirectors()
        showMenu()
    elif exit == 2:
        print('\nEnter the name of two films from the dataset')
        filmComparison()
        showMenu()
    elif exit == 3:
        analyseGrossEarnings()
        showMenu()
    elif exit == 4:
        generalAnalysis()
        showMenu()
    elif exit == 5:
        plotEarningsAndIMDBScores()
        showMenu()
    elif exit == 6:
        print('\nYou have now successfully exited from the menu.')
    else:
        print('\nInvalid Option! Please try again.\n')
        showMenu()    

# Show main menu
showMenu()   
```
